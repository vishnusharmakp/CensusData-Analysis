```python
#Import pandas, numpy and regular expression libraries
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt

#To get the data and also convert all "?" to NaN
set1 = pd.read_csv('Data-5300.csv',na_values = " ?")

#To check the dimension of table
set1.shape

#Print the data
set1
```

```python
#To check how many NaN values are present in the entire data having column as reference
set1.isnull().sum()
```

```python
#To print the number of rows before cleaning
row1 = set1.shape[0]
row1
```

```python
#To delete the rows having NaN values
set1.dropna(inplace=True)

#To read the data
set1
```

```python
#To check how many NaN values are present in the entire data after deleting
set1.isnull().sum()
```

```python
#To print the number of rows after cleaning
row2 = set1.shape[0]
row2
```

```python
#To check how many rows are deleted
row1-row2
```

```python
#To save the updated dataset into a new file
set1.to_csv('updated.csv', index=False)
```

```python
#To read the new updated data set
set2 = pd.read_csv('updated.csv')
set2
```

```python
#To get random sample and print
sample_data = set2.sample(125, random_state = 5121999)
sample_data
```

```python
#To create a new column to categorize the age
set2['Age-grp'] = set2.apply(lambda _: '', axis=1)
```

```python
#To classify the age group 
set2.loc[set2['Age']<=20, ['Age-grp']] = ['0-20']
set2.loc[set2['Age']>20, ['Age-grp']] = ['21-40']
set2.loc[set2['Age']>40, ['Age-grp']] = ['41-60']
set2.loc[set2['Age']>60, ['Age-grp']] = ['60+']
set2
```

```python
#To create a new column to categorize the employment 
set2['Employment'] = set2.apply(lambda _: '', axis=1)
```

```python
#Duplicate the column to make it easy
set2['Employment'] = set2['Work-Class']
```

```python
#Replace the new column as desired 
set2["Employment"] = set2["Employment"].str.replace("Self-emp-inc", "Self-emp")
set2["Employment"] = set2["Employment"].str.replace("Self-emp-not-inc", "Self-emp")
set2["Employment"] = set2["Employment"].str.replace("Local-gov", "Government")
set2["Employment"] = set2["Employment"].str.replace("State-gov", "Government")
set2["Employment"] = set2["Employment"].str.replace("Federal-gov", "Government")
set2
```

```python
#To plot the histogram for different age group
plt.figure(figsize=(10,8))
plt.title("Age Group Analysis")
sns.countplot(x='Age-grp',data=set2, order = ["0-20","21-40","41-60","60+"])
```

```python
#To count number of people under the same age group
counts = set2['Age-grp'].value_counts().to_dict()
print(counts)
```

```python
#To count number of people under the Employment type
counts = set2['Employment'].value_counts().to_dict()
print(counts)
```

```python
#To get the count of people under the same age group of different employment
set2.groupby(["Age-grp", "Employment"]).size()
```
